```python
# Convert 'Salary' column to numeric
salary_clean = pd.to_numeric(hitters_data['Salary'], errors='coerce')

# Compute the mean of the non-missing salary values (ignoring NaN)
smean = salary_clean.mean()
print("Mean of non-missing salary values:", smean)

# Replace NaN values with the mean and convert the entire 'Salary' column to integer
hitters_data['Salary'] = hitters_data['Salary'].fillna(smean).astype(float).astype(int)

# Display the updated 'Salary' column
print("Updated salary values (after replacing NaN values):\n", hitters_data['Salary'].head(10))
```

```python
# Measures of Central Tendency (Mean, Median, Mode)
runs_mean = hitters_data["Runs"].mean()
runs_median = hitters_data["Runs"].median()
runs_mode = hitters_data["Runs"].mode()[0]

salary_mean = hitters_data["Salary"].mean()
salary_median = hitters_data["Salary"].median()
salary_mode = hitters_data["Salary"].mode()[0]  # Getting the first mode value

print("Measures of Central Tendency:")
print(f"Runs - Mean: {runs_mean}, Median: {runs_median}, Mode: {runs_mode}")
print(f"Salary - Mean: {salary_mean}, Median: {salary_median}, Mode: {salary_mode}")
```

```python
# Measures of Dispersion (Standard Deviation, Variance)
runs_std_dev = hitters_data["Runs"].std()
runs_variance = hitters_data["Runs"].var()

salary_std_dev = hitters_data["Salary"].std()
salary_variance = hitters_data["Salary"].var()

print("\nMeasures of Dispersion:")
print(f"Runs - Standard Deviation: {runs_std_dev}, Variance: {runs_variance}")
print(f"Salary - Standard Deviation: {salary_std_dev}, Variance: {salary_variance}")
```

```python
# Quartile Ranges and IQR
runs_Q1 = hitters_data["Runs"].quantile(0.25)
runs_Q3 = hitters_data["Runs"].quantile(0.75)
runs_IQR = runs_Q3 - runs_Q1

salary_Q1 = hitters_data["Salary"].quantile(0.25)
salary_Q3 = hitters_data["Salary"].quantile(0.75)
salary_IQR = salary_Q3 - salary_Q1

print("\nQuartiles and IQR:")
print(f"Runs - Q1: {runs_Q1}, Q3: {runs_Q3}, IQR: {runs_IQR}")
print(f"Salary - Q1: {salary_Q1}, Q3: {salary_Q3}, IQR: {salary_IQR}")
```

```python
# Correlation
correlation = hitters_data["Runs"].corr(hitters_data["Salary"])

print("\nCorrelation between Runs and Salary:")
print(f"Correlation: {correlation}")
```

```python
# 1. Create a synthetic time variable (monthly)
hitters_data['Date'] = pd.date_range(start='2020-01-01', periods=len(hitters_data), freq='M')

# 2. Time-Based Indexing
time_series_data = hitters_data.set_index('Date')

# 3. Visualizing Time Series
plt.figure(figsize=(14, 6))
plt.plot(time_series_data['Salary'], label='Salary over Time', color='blue')
plt.title('Salary Time Series - 22BCE3775')
plt.xlabel('Date')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# 4. Rolling Time Series Data (e.g., 3-month rolling average)
time_series_data['SMA'] = time_series_data['Salary'].rolling(window=3).mean()

# Plotting Salary and SMA
plt.figure(figsize=(14, 6))
plt.plot(time_series_data['Salary'], label='Original Salary', color='blue')
plt.plot(time_series_data['SMA'], label='3-Month SMA', color='orange')
plt.title('Salary with 3-Month Simple Moving Average -  22BCE3775')
plt.xlabel('Date')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# 5. Resampling Time Series Data (e.g., yearly average salary)
yearly_salary = time_series_data['Salary'].resample('Y').mean()

# Plotting Resampled Data
plt.figure(figsize=(14, 6))
plt.plot(yearly_salary, label='Yearly Average Salary', color='green')
plt.title('Yearly Average Salary -  22BCE3775')
plt.xlabel('Year')
plt.ylabel('Average Salary')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose  # Import the seasonal_decompose function

# 6. Seasonal Decomposition of Time Series
decomposition = seasonal_decompose(time_series_data['Salary'], model='additive', period=12)  # Assuming monthly data
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Plotting the decomposed components
plt.figure(figsize=(14, 10))
plt.subplot(4, 1, 1)
plt.plot(time_series_data['Salary'], label='Original', color='blue')
plt.title('Original Time Series - 22BCE3775')
plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend', color='orange')
plt.title('Trend Component')
plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonal', color='green')
plt.title('Seasonal Component')
plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual', color='red')
plt.title('Residual Component')
plt.tight_layout()
plt.show()
```
