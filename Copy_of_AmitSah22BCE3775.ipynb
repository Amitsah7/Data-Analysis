```python
# Convert 'Salary' column to numeric
salary_clean = pd.to_numeric(hitters_data['Salary'], errors='coerce')

# Compute the mean of the non-missing salary values (ignoring NaN)
smean = salary_clean.mean()
print("Mean of non-missing salary values:", smean)

# Replace NaN values with the mean and convert the entire 'Salary' column to integer
hitters_data['Salary'] = hitters_data['Salary'].fillna(smean).astype(float).astype(int)

# Display the updated 'Salary' column
print("Updated salary values (after replacing NaN values):\n", hitters_data['Salary'].head(10))
```

```python
# Measures of Central Tendency (Mean, Median, Mode)
runs_mean = hitters_data["Runs"].mean()
runs_median = hitters_data["Runs"].median()
runs_mode = hitters_data["Runs"].mode()[0]

salary_mean = hitters_data["Salary"].mean()
salary_median = hitters_data["Salary"].median()
salary_mode = hitters_data["Salary"].mode()[0]  # Getting the first mode value

print("Measures of Central Tendency:")
print(f"Runs - Mean: {runs_mean}, Median: {runs_median}, Mode: {runs_mode}")
print(f"Salary - Mean: {salary_mean}, Median: {salary_median}, Mode: {salary_mode}")
```

```python
# Measures of Dispersion (Standard Deviation, Variance)
runs_std_dev = hitters_data["Runs"].std()
runs_variance = hitters_data["Runs"].var()

salary_std_dev = hitters_data["Salary"].std()
salary_variance = hitters_data["Salary"].var()

print("\nMeasures of Dispersion:")
print(f"Runs - Standard Deviation: {runs_std_dev}, Variance: {runs_variance}")
print(f"Salary - Standard Deviation: {salary_std_dev}, Variance: {salary_variance}")
```

```python
# Quartile Ranges and IQR
runs_Q1 = hitters_data["Runs"].quantile(0.25)
runs_Q3 = hitters_data["Runs"].quantile(0.75)
runs_IQR = runs_Q3 - runs_Q1

salary_Q1 = hitters_data["Salary"].quantile(0.25)
salary_Q3 = hitters_data["Salary"].quantile(0.75)
salary_IQR = salary_Q3 - salary_Q1

print("\nQuartiles and IQR:")
print(f"Runs - Q1: {runs_Q1}, Q3: {runs_Q3}, IQR: {runs_IQR}")
print(f"Salary - Q1: {salary_Q1}, Q3: {salary_Q3}, IQR: {salary_IQR}")
```

```python
# 1D Statistical Data analysis.
# A measure of Central tendency (All possibilities of mean, all possibilities of median - Quantile, deciles (ntile), percentiles)
# Measure of Dispersions (Range, IQR, Interdecile range, Deviation (mean and Standard deviation), Skewness and Kurtosis)

# Mean (Arithmetic, Geometric, Harmonic)
mean_value = hitters_data['AtBat'].mean()
geometric_mean = np.exp(np.mean(np.log(hitters_data['AtBat'][hitters_data['AtBat'] > 0])))  # Handle 0 or negative values
harmonic_mean = len(hitters_data['AtBat']) / np.sum(1.0 / hitters_data['AtBat'][hitters_data['AtBat'] > 0])  # Handle 0 or negative values

# Median (including Quantiles, Deciles, Percentiles)
median_value = hitters_data['AtBat'].median()
quantiles = hitters_data['AtBat'].quantile([0.25, 0.5, 0.75])  # Q1, Q2 (Median), Q3
deciles = hitters_data['AtBat'].quantile(np.arange(0.1, 1, 0.1))  # Deciles (10%, 20%, ..., 90%)
percentiles = hitters_data['AtBat'].quantile([0.01, 0.05, 0.95, 0.99])  # Percentiles (1%, 5%, 95%, 99%)

# 2. Measures of Dispersion
range_value = hitters_data['AtBat'].max() - hitters_data['AtBat'].min()
iqr_value = hitters_data['AtBat'].quantile(0.75) - hitters_data['AtBat'].quantile(0.25)
mean_deviation = np.mean(np.abs(hitters_data['AtBat'] - mean_value))
std_deviation = hitters_data['AtBat'].std()

# Skewness and Kurtosis
skew_value = skew(hitters_data['AtBat'], nan_policy='omit')
kurtosis_value = kurtosis(hitters_data['AtBat'], nan_policy='omit')

# Print results
print("Mean: ", mean_value)
print("Geometric Mean: ", geometric_mean)
print("Harmonic Mean: ", harmonic_mean)
print("Median: ", median_value)
print("Quantiles: ", quantiles)
print("Deciles: ", deciles)
print("Percentiles: ", percentiles)
print("Range: ", range_value)
print("IQR: ", iqr_value)
print("Mean Deviation: ", mean_deviation)
print("Standard Deviation: ", std_deviation)
print("Skewness: ", skew_value)
print("Kurtosis: ", kurtosis_value)
```

```python
# 2D Analysis

#  2-way contingency table
# Categorical-Categorical
contingency_cat_cat = pd.crosstab(hitters_data['League'], hitters_data['Division'])
print(contingency_cat_cat)
print("\n")
# Numerical-Numerical
correlation = hitters_data[['AtBat', 'Hits']].corr()
print(correlation)
print("\n")
# Categorical-Numerical
summary_by_league = hitters_data.groupby('League')['AtBat'].describe()
print(summary_by_league)
```

```python
# 3-way Contingency Table
contingency_3way = pd.crosstab([hitters_data['League'], hitters_data['Division']], hitters_data['Years'])
print(contingency_3way)
```

```python
# Row Profile, Column Profile, Relative Frequency Dataset, Chi-Square Dataset
# Row profile
row_profile = contingency_cat_cat.div(contingency_cat_cat.sum(axis=1), axis=0)
print(row_profile)
print("\n")
# Column profile
column_profile = contingency_cat_cat.div(contingency_cat_cat.sum(axis=0), axis=1)
print(column_profile)
print("\n")
# Relative frequency (Overall)
relative_freq = contingency_cat_cat.div(contingency_cat_cat.sum().sum())
print(relative_freq)
print("\n")
# Chi-Square test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_cat_cat)
print(f'Chi2 Statistic: {chi2}, p-value: {p}')
```

```python
# K-means Clustering

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# numerical columns for clustering
data = hitters_data[['AtBat', 'Hits']]

# Standardizing the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Using the elbow method to find the optimal number of clusters
wcss = []  # Within-cluster sum of squares (WCSS)

# Looping through a range of possible cluster numbers (1 to 10)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the elbow graph
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal Number of Clusters 22BCE3775')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(data_scaled)

# Getting the cluster labels
clusters = kmeans.labels_

# Adding cluster labels to the dataset
hitters_data['Cluster'] = clusters

# Display the sizes of each cluster
cluster_sizes = np.bincount(kmeans.labels_)
print("Cluster Sizes:", cluster_sizes)
print("\n")
# Display cluster centers (centroid vectors)
print("Cluster Centers (Centroids):\n", kmeans.cluster_centers_)
print("\n")
# Display all available components of the KMeans model
print("Available Components in KMeans Model:\n", dir(kmeans))

```

```python
# Visualizing the clusters
plt.figure(figsize=(8,6))

# Plotting each cluster with a different color
plt.scatter(data_scaled[clusters == 0, 0], data_scaled[clusters == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(data_scaled[clusters == 1, 0], data_scaled[clusters == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(data_scaled[clusters == 2, 0], data_scaled[clusters == 2, 1], s=100, c='green', label='Cluster 3')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

# Adding titles and labels
plt.title('Clusters of Hitters (K-Means) 22BCE3775')
plt.xlabel('AtBat')
plt.ylabel('Hits')
plt.legend()
plt.show()
```

```python
# Principle Component Analysis

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select only numeric columns for PCA
numeric_data = hitters_data.select_dtypes(include=['float64', 'int64']).dropna()

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Perform PCA with 5 components to explore
pca = PCA(n_components=5)
pca_result = pca.fit_transform(scaled_data)

# Print PCA results
print("PCA - Explained Variance (Eigenvalues):")
print(pca.explained_variance_)

print("\nPCA - Component Loadings:")
print(pca.components_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], edgecolor='k')
plt.title("PCA - First Two Components")
plt.xlabel(f"PC 1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC 2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.show()
```

```python
# Biplot of the first two principal components
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], marker='o', s=100, color='b')

# Add arrows for component loadings
for i in range(numeric_data.shape[1]):
    plt.arrow(0, 0, pca.components_[0, i] * max(pca_result[:, 0]),
              pca.components_[1, i] * max(pca_result[:, 1]),
              color='r', alpha=0.5, head_width=0.05, head_length=0.1)
    plt.text(pca.components_[0, i] * max(pca_result[:, 0]),
             pca.components_[1, i] * max(pca_result[:, 1]),
             numeric_data.columns[i], color='r', ha='center', va='center')

plt.title("Biplot of PCA Components")
plt.xlabel(f"PC 1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC 2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.show()
```

```python
# Scree plot to determine the optimal number of components
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title("Scree Plot")
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.xticks(range(1, len(pca.explained_variance_ratio_)+1))
plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Perform PCA with the desired number of components
n_comp = 3  #  we're using 3 components based on the scree plot
pca = PCA(n_components=n_comp)
pca_result = pca.fit_transform(scaled_data)

# Biplot for both individuals and variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], marker='o', s=100, color='b', label='Individuals')

# Add arrows for component loadings (variables)
for i in range(numeric_data.shape[1]):
    plt.arrow(0, 0, pca.components_[0, i] * max(pca_result[:, 0]),
              pca.components_[1, i] * max(pca_result[:, 1]),
              color='r', alpha=0.5, head_width=0.05, head_length=0.1)
    plt.text(pca.components_[0, i] * max(pca_result[:, 0]),
             pca.components_[1, i] * max(pca_result[:, 1]),
             numeric_data.columns[i], color='r', ha='center', va='center')

# Add labels and title
plt.title("PCA - Biplot (Individuals and Variables)")
plt.xlabel(f"PC 1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC 2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.legend()
plt.grid(True)
plt.show()
```
